```python
#create categorical comparision

sns.catplot(x="parental level of education", y="math score", hue="gender", kind="point", data=df,
             height=8, aspect=1)
```

```python
#how to find the line correlation between two numeric data
#here we chech correlation between writing and math score
plt.figure(figsize=(15,5))

# Add title
plt.title("the correlaton between writing score and math score")

# Line chart showing daily global streams of 'Shape of You'
sns.lineplot(data=df['writing score'], label="writing score")

# Line chart showing daily global streams of 'Despacito'
sns.lineplot(data=df['math score'], label="math score")

# Add label for horizontal axis
plt.xlabel("score")
```

```python
sns.color_palette("Paired")
sns.regplot(x='writing score', y='math score', data=df)
```

```python
sns.lmplot(x="writing score", y="math score", data=df)
```

```python
sns.lmplot(x="writing score", y="reading score", data=df)
```

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Average reading score by race ethicity")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(y=df['reading score'], x=df['race/ethnicity'])

# Add label for vertical axis
plt.ylabel(" reading score ")
```

```python
#to compare to different groups in a great way!
sns.catplot(x="race/ethnicity", y="reading score", hue="gender",
            col="lunch", aspect=.7,
            kind="swarm", data=df)
```

```python
#create a heatmap
# Set the width and height of the figure
plt.figure(figsize=(14,7))
cols=['math score','reading score']
new_data=df.loc[:,cols]

# Add title
plt.title("heat map of reading and math score")

# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=new_data, annot=True)

# Add label for horizontal axis
plt.xlabel("Heat map")
# as you can see we have a lots of data which makes the  understanding of data impossible ,
#so it's better to choose a special part!!!!
```
